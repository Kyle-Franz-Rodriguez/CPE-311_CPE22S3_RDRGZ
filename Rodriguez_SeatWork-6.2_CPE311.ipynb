Filtering with Boolean Masks

```python
# Using AND
(meteorite['mass (g)'] > 50) & (meteorite.fall == 'Found')
# This is output only True or False
```

```python
# Using the code above, we can filter the dataframe to return only the True rows
meteorite[(meteorite['mass (g)'] > 50) & (meteorite.fall == 'Found')]
```

```python
# We can use query() alternatively
meteorite.query("`mass (g)` > 50 and fall == 'Found'") # Use `` instead of ''
```

Calculating Summary Statistics

```python
# How many of the meteorites were found versus observed falling?
meteorite.fall.value_counts() # value_counts() just count each unique value inside 1 column
```

```python
# What proportion of meteorites are...
# To answer the question: check the number of unique values of more than 1 subset
meteorite.value_counts(subset=['nametype','fall'])
```

```python
meteorite.value_counts(subset=['nametype','fall'], normalize = True) # Normalize to see the distribution value of Fall
```

```python
meteorite.value_counts(subset=['nametype','fall'], normalize = False)
```
