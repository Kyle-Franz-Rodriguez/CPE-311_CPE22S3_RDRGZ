# Exercise 1

We want to look at data for the Facebook, Apple, Amazon, Netflix, and Google (FAANG) stocks, but we were given each as a separate CSV file. Combine them into a single file and store the dataframe of the FAANG data as faang for the rest of the exercises

```python
# 1. Read all csv files
import pandas as pd

dat = ['aapl.csv', 'amzn.csv', 'fb.csv', 'goog.csv', 'nflx.csv']
stock = [pd.read_csv(i) for i in dat]
```

```python
# 2. Add a column to each dataframe called 'ticker'
for i in range(len(stock)):
  stock[i] = stock[i].assign(ticker = dat[i].removesuffix('.csv').upper())
```

```python
# 3. Appemd all dataframes in a single dataframe
faang = pd.concat(stock)
faang
```

```python
# 4. Save the result in a csv file called faang.csv
faang.to_csv('faang.csv')
```

```python
# 1. Using web scraping,
# search for the list of the hospitals, their address and contact information. 
# Save the list in a new csv file, hospitals.csv
from bs4 import BeautifulSoup
import requests
```

```python
url = requests.get('https://shop.philcare.com.ph/accredited-hospitals')
soup = BeautifulSoup(url.text, 'html')
```

```python
# Finding the table
# I also inspected the website to find this
hospitals_text = soup.find_all('table')[0]
```

```python
# All column headers
headers_text = hospitals_text.find_all('th')
headers_text
```

```python
# Put headers in list
hospitals_headers_text = [i.text for i in headers_text] 
print(hospitals_headers_text)
```

```python
# Finding the data inside each column
hospitals_data_text = hospitals_text.find_all('tr')
```

```python
# Converting to dataframe
import pandas as pd

hospitals = pd.DataFrame(columns = hospitals_headers_text)
hospitals
```

```python
# Put data in list and append it to the dataframe
for tr in hospitals_data_text[1:]: # There was an error without this slicing
    row_dat = tr.find_all('td')
    hospitals_data = [data.text for data in row_dat]
    length = len(hospitals) # Check for length each iteration
    hospitals.loc[length] = hospitals_data # Append the list of rows based on the current length
```

```python
hospitals.head()
```

```python
# Cleaning the data
# Since we only want address and contact information, we just drop all other columns
# We only want Provider Name, Complete Address, and Contact No.
hospitals = hospitals.drop(columns = ['City', 'Province', 'Region', 'Area'])
hospitals.columns
```

```python
# For finalization, I could drop all of the NULL values
# I should replace emptry strings with NaN values using numpy
from numpy import nan
hospitals.replace('', nan, inplace = True)
hospitals.head()
```

```python
# Dropping all rows with NaN values
hospitals.dropna(inplace = True)
hospitals.head()
```

```python
# Save to csv
hospitals.to_csv('hospitals.csv')
```
