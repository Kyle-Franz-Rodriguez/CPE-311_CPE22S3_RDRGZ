# Section 2: Data Warngling

```python
taxis.head()
```

```python
mask = taxis.columns.str.contains('id$|store_and_fwd_flag', regex = True)
columns_to_drop = taxis.columns[mask]
columns_to_drop
```

```python
taxis.drop(columns = columns_to_drop) # No variable to store, hence not saving
taxis.head() # Notice that showing the dataframe doesn't save anything, it is because we do not save it in the line above.
```

```python
taxis_drop = taxis.drop(columns = columns_to_drop) # Saved
taxis.head()
```

```python
taxis = taxis.rename(columns = {'tpep_pickup_datetime': 'pickup', 'tpep_dropoff_datetime': 'dropoff'}) # Rename columns
taxis.columns
```

Change data type of a column

```python
taxis.dtypes # We want to change pickup and dropoff column.
# Note that the collumn tpep_pickup_datetime and tpep_dropoff_datetime wasn't renamed because I executed this code before the line above. Just sayin'
```

```python
# Change data type of a column
test = pd.to_datetime # This is a value that we want to apply. Notice the object output below
test
```

```python
taxis[['pickup', 'dropoff']] = taxis[['pickup','dropoff']].apply(test) # Applied dtype change
taxis.dtypes
```

Create new columns

```python
# This is quite similar in adding new items in a dictionary
taxis = taxis.assign(
    elapsed_time = lambda x: x.dropoff - x.pickup,
    cost_before_tip = lambda x: x.total_amount - x.tip_amount,
    tip_pct = lambda x: x.tip_amount / x.cost_before_tip,
    fees = lambda x: x.cost_before_tip - x.fare_amount,
    avg_speed = lambda x: x.trip_distance.div(
        x.elapsed_time.dt.total_seconds() / 60 / 60
    )
)
taxis.head(1)
# All newly added columns was always in the right side of the dataframe.
```

# Exercise 3

Using the meteorite data from the Meteorite_Landing.csv file, update the year column to only contain the year, convert it to a numeric data type, and create a new column indicating whether the meteorite was observed falling before 1970. Set the index to the id column and extract all the rows with IDs between 10,036 and 10,040 (inclusive) with loc[].

```python
# Read the meteorite data from the Meteorite_Landing.csv file
exer3_df = pd.read_csv('Meteorite_Landings.csv')
exer3_df.head()
```

```python
# Update the year column to only contain the year.
exer3_df['year'] = exer3_df['year'].str.slice(6, 11)
exer3_df.head()
```

```python
# Convert the year to numeric data type.
exer3_df['year'] = pd.to_numeric(exer3_df['year'])
exer3_df.dtypes
```

```python
# Create a new column indicating whether the meteorite was observed falling before 1970.
exer3_df = exer3_df.assign(Before1970 = lambda x: x.year < 1970)
exer3_df.head()
```

```python
# Set the index to the id column
exer3_df = exer3_df.set_index('id')
exer3_df.head()
```

```python
# Extract all the rows with IDs between 10,036 and 10,040 (inclusive) with loc[]
exer3_df = exer3_df.sort_index() # Sort index
exer3_df.loc[10036:10040]
```
