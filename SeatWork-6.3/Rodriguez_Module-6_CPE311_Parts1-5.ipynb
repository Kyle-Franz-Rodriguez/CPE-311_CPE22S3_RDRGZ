# Section 2: Data Warngling

```python
import pandas as pd
taxis = pd.read_csv('2019_Yellow_Taxi_Trip_Data.csv')
taxis.head()
```

```python
# Drop all columns with names starting with 'id' OR specific column name store_and_fwd_flag
mask = taxis.columns.str.contains('id$|store_and_fwd_flag', regex = True) # Use regex to read the $ sign.
columns_to_drop = taxis.columns[mask]
columns_to_drop
# We can do this in one line, but it is more readable this way
```

```python
taxis.drop(columns = columns_to_drop) # What if we didn't store this to a variable
taxis.head() # Notice that the displayed variable wasn't changed, it is because we do not save it in the line above.
```

```python
taxis_drop = taxis.drop(columns = columns_to_drop) # So always remember to save, or use inplace if possible
taxis_drop.head()
```

Change column names

```python
# In changing column name, we can use dataframe.rename()
# We have to store the columns we want to change together with its new name in a dictionary
new_column = {'tpep_pickup_datetime': 'pickup', 'tpep_dropoff_datetime': 'dropoff'}
taxis = taxis.rename(columns = new_column) # We can do this in one line, but it is more readable this way
taxis.columns
```

Change data type of a column

```python
taxis.dtypes # We want to change pickup and dropoff column.
```

```python
# Change data type of a column using dataframe.apply
test = pd.to_datetime # This is a value that we want to apply. Notice the object output below
test
```

```python
taxis[['pickup', 'dropoff']] = taxis[['pickup','dropoff']].apply(test) # Applied dtype change
taxis.dtypes
# In using dataframe.apply, we use an object to pass its value to every value in the extracted columns.
# Note that dataframe.apply changes the values of a column. We can also use it to change the datatype of a column
```

```python
# Anotherway of changing column datatype
taxis[['pickup', 'dropoff']] = pd.to_datetime(taxis[['pickup','dropoff']])
```

Create new columns

```python
# Creating new columns in a dataframe is quite similar in appending items in a dictionary
taxis['new_column'] = ['values'] # Just make sure you are adding appropriate value
```

```python
# Additionally, we can also use dataframe.assign if we want to save time adding many columns
taxis = taxis.assign(
    elapsed_time = lambda x: x.dropoff - x.pickup,
    cost_before_tip = lambda x: x.total_amount - x.tip_amount,
    tip_pct = lambda x: x.tip_amount / x.cost_before_tip,
    fees = lambda x: x.cost_before_tip - x.fare_amount,
    avg_speed = lambda x: x.trip_distance.div(
        x.elapsed_time.dt.total_seconds() / 60 / 60
    )
)
taxis.head(1)
# All newly added columns was always in the right side of the dataframe.
```

# Section 4: Reshaping data

If a dataset is not in a proper format for an analysis, we can reshape it to make it possible. 

```python
import pandas as pd
tsa = pd.read_csv('tsa_passenger_throughput.csv')
tsa.head()
```

```python
# If we want to lowercase the column names and rename it to only have the first word
# We can use lower() and split()
tsa = tsa.rename(columns = lambda x: x.lower().split()[0])
# split() separates strings with whitespaces to a list
# We can take the first element in the split() as if it were a list split()[index]
tsa.head()
```

Reshaping the dataset into desired number of columns

```python
# Transform dataset into long format by melting
# Melting can help convert our data into long format
tsa_melted = tsa.melt(
    id_vars = 'date', # Column that uniquely identifies a row (can be multiple)
    var_name = 'year', # Name for the new column created by melting
    value_name = 'travelers' # Name for the new column containing values from melted columns
)
tsa_melted.head()
# Basically what this does is it transforms columns into a row and passes their value into a new column
```

```python
# We can select random entries in a dataframe using dataframe.sample()
n = 5 # n number of entries
tsa_melted.sample(n, random_state = 1)
# Notice in the output we have an error where the date and year doesn't match
```

Convert the tsa_melted into time series

```python
tsa_melted.date = pd.to_datetime(tsa_melted['date']) # convert date column to datetime
tsa_melted.dtypes
```

```python
tsa_melted = tsa_melted.assign(
    date = lambda x: pd.to_datetime(x.year + x.date.dt.strftime('-%m-%d'))
) # This changes the year of the date to match the column year
tsa_melted.sample(5, random_state = 1)
```

Dropping null values

```python
# Notice that we have some null values
tsa_melted.sort_values('date').tail()
```

```python
# We can use dataframe.dropna() to drop the null values
tsa_melted = tsa_melted.dropna()
tsa_melted.sort_values('date').tail()
```

Pivoting

```python
# Pivoting means reshaping the dataframe, organizing it differently
tsa_pivot = tsa_melted.pivot(index = 'date', columns = 'year', values = 'travelers')
tsa_pivot
# We can use this to answer a question:
```

```python
# Compare TSA traveler throughput on March 10 across the years
# To answer this we could do:

# Extract the 3rd month and days from 1 to 10
# Make a new column with the day values
# Pivot the new column
tsa_march = tsa_melted\
  .query('date.dt.month == 3 and date.dt.day <=10')\
  .assign(day_in_march = lambda x: x.date.dt.day)\
  .pivot(index = 'year', columns = 'day_in_march', values = 'travelers') 
tsa_march
```

Transposing

```python
# We can flip rows and columns using dataframe.T
tsa_march.T
```

Merging

```python
# We can merge two dataframes using dataframe.merge
# Let's say we want to add more information about TSA throughput
# The information we want to add is holiday and it is present in a different dataframe
holidays = pd.read_csv('holidays.csv', parse_dates = True, index_col = 'date')
holidays.loc['2019']
```

```python
# We can merge this dataframe with the TSA traveler throughput data
# Doing so will provide more context for our analysis
tsa_melted_holidays = tsa_melted\
  .merge(holidays, left_on = 'date', right_index = True, how = 'left')\
  .sort_values('date')
tsa_melted_holidays.head()
```

We can take this a step further by marking a few days before and after each holiday as part of the holiday. This would make it easier to compare holiday travel across years and look for any uptick in travel around the holidays:

```python
tsa_melted_holiday_travel = tsa_melted_holidays.assign(
    holiday = lambda x: x.holiday.ffill(limit = 1).bfill(limit = 2)
)
tsa_melted_holiday_travel.head()
```

Aggregation and grouping

We can perform aggregations to summarize our cleaned data

Pivot tables

```python
# We can build a pivot table to sum-up column values and rearrange the dataset like before
# This is similar to groupby but with more functionality
tsa_melted_holiday_travel.pivot_table(
    index = 'year', columns = 'holiday', sort = False, values = 'travelers', aggfunc = 'sum'
) # This reshapes the data and sums up all of the values per index
```

```python
# We can also use dataframe.pct_change() to see the biggest change
tsa_melted_holiday_travel.pivot_table(
    index = 'year', columns = 'holiday', sort = False, values = 'travelers', aggfunc = 'sum'
).pct_change(fill_method = None)
```

Formatting improvements

```python
# We can set a display option for all floats
pd.set_option('display.float_format', '{:,.0f}'.format)
```

```python
# One last pivot table
# We can group Christmas Eve and Christmas Day together
# We can also group New Year's Eve and New Year's Day together
import numpy as np

pivot_table = tsa_melted_holiday_travel.assign(
    year = lambda x: np.where(
        x.holiday == "New Year's Day", pd.to_numeric(x.year) - 1, x.year
    ).astype(str),
    holiday = lambda x: np.where(
        x.holiday.str.contains('Christmas|New Year', regex = True),
        x.holiday.str.replace('Day|Eve', '', regex = True).str.strip(),
        x.holiday
    )
).pivot_table(
    index = 'year', columns = 'holiday', sort = False,
    values = 'travelers', aggfunc = 'sum',
    margins = True, margins_name = 'Total'
)

# reorder columns by order in the year
pivot_table.insert(5, "New Year's", pivot_table.pop("New Year's"))
pivot_table
```
