```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
print(salaries)
```

Using the data generated above, calculate the following statistics without importing anything from the statistics module in the
standard library (https://docs.python.org/3/library/statistics.html) and then confirm your results match up to those that are
obtained when using the statistics module (where possible):



*   Mean
*   Median
*   Mode
*   Sample Variance
*   Sample Standard Deviation



```python
# Mean
def summation(nlist): # To count for summation to be used for the mean
  if len(nlist) == 1: return nlist
  else: return nlist[0] + sum(nlist[1:])

def mean(nlist):
  return (summation(nlist) / len(nlist))

mean(salaries)
```

```python
# Median

def median(nlist):
  nlist = sorted(nlist)
  n = len(nlist)
  if n % 2 == 0: return ((nlist[int(n/2)] + nlist[int(n/2)+1]) / 2)
  else: return  nlist[(n+1)/2]

median(salaries)
```

```python
# Mode
def mode(nlist):
  storage = {}
  for i in nlist: # This loop is for counting the frequency of each unique values in the list
    if not i in storage: storage[i] = 1 # The iterated key is new and wasn't found.
    else: storage[i] += 1 # Increment the value if the key was already found.
  return max(storage, key = storage.get)

mode(salaries)
```

```python
# Sample Variance
def variance_sample(nlist):
  storage = 0 # For the summation
  for i in nlist: storage += ((i - mean(nlist))**2) # Based on formula of sample variance
  return (storage / (len(nlist) - 1))

variance_sample(salaries)
```

```python
# Sample Standard Deviation
import math # For square root

def standard_deviation_sample(nlist):
  return math.sqrt(variance_sample(nlist))

standard_deviation_sample(salaries)
```

# Exercise 2

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:

*   Range
*   Coefficient of variation interquartile range
*   Quartile coefficient of dispersion

```python
# Range
def st_range(nlist):
  return (max(nlist) - min(nlist))

st_range(salaries)
```

```python
# Coefficient of variation interquartile range
# Calculated using IQR / Median
def cv_iqr(nlist):
  from statistics import median, quantiles
  nlist = sorted(nlist)
  quartile = quantiles(nlist, n=4) # For quartiles (0th and 2nd index is the q1 and q3)
  return 100 * ((quartile[2] - quartile[0]) / median(nlist))

print(f"{cv_iqr(salaries)} %")
```

```python
# Quartile coefficient of dispersion
# Calculated using IQR / Q3 + Q1
def qc_dispersion(nlist):
  from statistics import quantiles
  nlist = sorted(nlist)
  quartile = quantiles(nlist, n=4) # For quartiles
  return ((quartile[2] - quartile[0]) / (quartile[2] + quartile[0]))

qc_dispersion(salaries)
```
